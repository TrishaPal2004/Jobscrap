```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

# Set the correct path to your ChromeDriver
chrome_driver_path = r'C:\Users\trish\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe'

# Create a Service object
service = Service(chrome_driver_path)

# Launch Chrome using the Service object
driver = webdriver.Chrome(service=service)

# Open the LinkedIn Jobs URL
url1 = "https://www.linkedin.com/jobs/search/?currentJobId=4197308361&f_WT=2&geoId=102713980&keywords=software%20engineer%20intern&origin=JOB_SEARCH_PAGE_SEARCH_BUTTON&refresh=true"
driver.get(url1)
driver.implicitly_wait(10)  # Wait for elements to load

```

```python
from selenium.webdriver.common.by import By

# Find elements correctly
elements = driver.find_elements(By.CLASS_NAME, 'results-context-header__job-count')

# Ensure there is at least one element before accessing index 0
if elements:
    n = elements[0]
    print(n.text)  # Print the job count text
else:
    print("No elements found with the given class name.")
```

```python
y=n.text
y=pd.to_numeric(y.replace(",", ""), errors='coerce')
y
```

```python
from selenium.webdriver.common.by import By

# Ensure y is an integer
y = int(y) if isinstance(y, (str, float)) else y  # Convert safely if needed

company_name=[]
# Find all company elements
companies = driver.find_elements(By.CLASS_NAME, 'base-search-card__subtitle')

# Loop through available elements (ensure `y` is within range)
for j in range(min(y, len(companies))):  
    company_name.append(companies[j].text)
companyfinal=pd.DataFrame(company_name,columns=['companies'])
print(companyfinal)
```

```python
from selenium.webdriver.common.by import By

# Ensure y is an integer
y = int(y) if isinstance(y, (str, float)) else y  # Convert safely if needed

title_name=[]

# Find all company elements
titles = driver.find_elements(By.CLASS_NAME, 'base-search-card__title')

# Loop through available elements (ensure `y` is within range)
for j in range(min(y, len(titles))):  
    title_name.append(titles[j].text)
titlefinal=pd.DataFrame(title_name,columns=['titles'])
print(titlefinal)
```

```python
from selenium.webdriver.common.by import By

# Ensure y is an integer
y = int(y) if isinstance(y, (str, float)) else y  # Convert safely if needed

location=[]

# Find all company elements
locations = driver.find_elements(By.CLASS_NAME, 'job-search-card__location')

# Loop through available elements (ensure `y` is within range)
for j in range(min(y, len(locations))):  
    location.append(locations[j].text)
locationfinal=pd.DataFrame(location,columns=['location'])
print(locationfinal)
```

```python
final=companyfinal.join(titlefinal)
final=final.join(locationfinal)
```
